#### * To get the next column "Total weight as per X (KG)", we need to follow the following steps:
* 1. Merging the two tables Order Report with SKU Master order
* 2. This will lead to mapping order ID's , SKU's and weight in grams
* 3. Finding weight as per the no. of orderss and thus gaining final weight
* 4. Also since the Order Report have multiple entries with same SKU's, the merging will take all those entries.
* 5. Further bringing them as per Order ID, by using Group By function

```python
merge_SKU=orderrep.merge(skumaster,on='SKU',how='left')
```

```python
merge_SKU
```

```python
merge_SKU['Tot Wt']=merge_SKU['Order Qty']*merge_SKU['Weight (g)'].copy()
```

```python
# Grouping the SKU's wrt to Order No
merge_SKU.groupby('ExternOrderNo')['Tot Wt'].sum().to_frame()
```

```python
data1=data1.merge(merge_SKU.groupby('ExternOrderNo')['Tot Wt'].sum().to_frame(),left_on='OrderID',right_on='ExternOrderNo')
```
